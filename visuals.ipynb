```python
import matplotlib.pyplot as plt
```

```python
categories = ["A", "B", "C", "D"]
values = [24, 46, 13, 38]
max_value = max(values)

#colors
colors = ["green" if value != max_value else "purple" for value in values]
plt.bar(categories, values, color=colors, width=0.8,bottom=0)
plt.title("Bar Chart: Compare Categories", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"})
plt.xlabel("Category" , fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"})
plt.ylabel("Value" , fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"})
plt.show()




```

```python
import matplotlib.pyplot as plt

Domain = ["AI", "ML", "DS", "Python"]
Cgpa = [8.0, 7.5, 9.0, 6.9]

colors = ["blue" if value > 8.5 else "red" for value in Cgpa]

plt.figure(figsize=(8,5))
plt.bar(Domain, Cgpa, color=colors)


plt.title("CGPA vs Domain")
plt.xlabel("Domain")
plt.ylabel("CGPA")
plt.show()
```

# Pie Chart

```python
labels = ["Product A", "Product B", "Product C", "Product D"]
sizes = [38,25,20,20]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("PIE CHART")
plt.show
```

Histogram

```python
data = np.random.normal(loc=60, scale=10, size=1000)

plt.hist(data, bins=30, color="green", edgecolor="white")
plt.title("HISTOGRAM")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

SCATTER PLOT : shows relationship bw two var

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [35,42,48,55,65,72,80,90]

plt.scatter(study_hours, marks, color="red", edgecolors="white")

plt.title("Study hrs vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid()
plt.show()
```
