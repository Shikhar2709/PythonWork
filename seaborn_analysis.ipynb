```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#check list of fnlity of seaborn
print(dir(sns))
```

```python
ft =sns.load_dataset("flights")
```

```python
ft
```

```python
sns.barplot(data=ft, x="month", y="passengers")
plt.xticks(rotation=45)
plt.title("Passenger Distribution")
plt.show()
```

```python
sns.violinplot(data=ft, x="month", y="passengers", color='red')
plt.xticks(rotation=45)
plt.title("Passenger Distribution")
plt.show()
```

```python
sns.scatterplot(data=ft, x="year", y="passengers")
plt.title("Passenger Vs Year")
plt.show()
```

```python
sns.histplot(ft["passengers"], bins=10)
plt.title("Passenger Distribution")
plt.show()
```

```python
#kde plot : only passenger
sns.kdeplot(ft["passengers"], fill=True)
plt.title("Density Plot")
plt.show()
```

```python
#heatmap
pivot = ft.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot, cmap="YlGnBu")
plt.title("flights heatmap")
plt.show()
```

```python
#pair plot
sns.pairplot(ft)
plt.show()
```

# CountPlot

```python
sns.countplot(data=ft, x="month")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,8))
#line plot
plt.subplot(2,2,1)
sns.lineplot(data=ft, x="year", y="passengers")
plt.show()
```

```python
#line plot with hue
sns.lineplot(data=ft, x="year", y="passengers", hue="month")
plt.title("Monthly Passenger Trend")
plt.show()
```

```python
#relational plot :year passenger
month_order = [
    "Jan", "Feb", "Mar", "Apr", "May","June"
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"

]
sns.relplot(
    data=ft,
    x="year",
    y="passengers",
    kind="line",
    hue="month",
    legend="auto"
)
plt.show()
```
