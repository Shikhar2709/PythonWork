```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset 
df = pd.read_csv('E:/Collge _Projects/PythonWork/numpypr/data.csv', encoding='latin-1')
```

```python
df.dropna(subset=["Description"])
```

```python
df.dropna(subset=["CustomerID"])
```

```python
# if there data is missing customer id we suggest guest trans
df_with_customer = df.dropna(subset=["CustomerID"]).copy()
df_with_customer["CustomerID"] = df_with_customer["CustomerID"].astype("int")
print(df_with_customer)
```

```python
df_with_customer['CustomerID']
```

```python
#check iscancelled payment and invoice no  and check data type confirm false cancelled truue 

df["IsCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C").sum()
```

```python
df["IsCancelled"]
```

```python
# create new df only keeping valid trans

df_clean = df[
    (~df["IsCancelled"]) & (df["Quantity"] > 0) & (df["UnitPrice"] > 0)
].copy() # separate copy of the filtered data frame
```

```python
df_clean
```

```python
#convert the invoice date from string to datetime format
#this allow us to do time series analysis and extract features like month, day, hour, etc.

df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"], format="%m/%d/%Y %H:%M")
```

```python
df_clean["InvoiceDate"]
```

```python
# create a new column TotalPrice by multiplying Quantity and UnitPrice using numpy's multiply function
#total price = quantity * unit price
df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"], df_clean["UnitPrice"])
```

```python
df_clean["TotalPrice"]
```

```python
print("STEP3: Data Cleaned")
print("Orignal rows:", df.shape[0], "Cleaned rows:", df_clean.shape[0])
print("Cancelled Order removed:", df["IsCancelled"].sum())
print(df_clean[["InvoiceNo", "Quantity", "UnitPrice", "TotalPrice", "InvoiceDate"]].head())
```

# Step 4 Date based feature for time analysis

```python
# extract year from invoice date and create a new column Year
df_clean["Year"] = df_clean["InvoiceDate"].dt.year
df_clean["Month"] = df_clean["InvoiceDate"].dt.month
df_clean["Day"] = df_clean["InvoiceDate"].dt.day
df_clean["Hour"] = df_clean["InvoiceDate"].dt.hour
#extract the name of the day of week
df_clean["DayOfWeek"] = df_clean["InvoiceDate"].dt.day_name()
```

```python
print("STEP4: Feature Extraction")
print(df_clean[["InvoiceDate", "Year", "Month", "Day", "Hour", "DayOfWeek"]].head(10))
print("\n")
```

# Step 5 : key bussiness metrics

```python
total_revenue = np.sum(df_clean["TotalPrice"])
```

```python
total_orders = df_clean["InvoiceNo"].nunique()
```

```python
total_customers = df_clean["CustomerID"].nunique()
```

```python
avg_order_value = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().mean()
```

```python
print("Step 5: key metrics")
print(f"Total revenue: {total_revenue:.2f}")
print(f"Total orders: {total_orders}")
print(f"Total customers: {total_customers}")
print(f"Average order value: {avg_order_value:.2f}")
```

# STEP 6 : (by revenue and by quantity)

```python
# find 10 product names with the highest total revenue
top_products_revenue =(
 df_clean.groupby("Description")["TotalPrice"].sum().sort_values(ascending=False).head(10)
)
#find 10 product names with the highest total quantity sold
top_products_quantity = (
    df_clean.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
)
```

```python
#display results
print("Top 10 products by revenue:")
print(top_products_revenue)
print("\nTop 10 products by quantity:")
print(top_products_quantity)
```

# Step 7 Country wise revenue analysis

```python
# cal total revenue gen by each country and sort the result in descending order
country_revenue = (
    df_clean.groupby("Country")["TotalPrice"].sum().astype(int).sort_values(ascending=False)
)
```

```python
#display results
print("Step 7: Revenue by country (top10)")
print(country_revenue.head(10))
print("\n")
```

# Step 8: Monthly Sales trend

```python
df_clean["YearMonth"] = df_clean["InvoiceDate"].dt.to_period("M")
monthly_sales = df_clean.groupby("YearMonth")["TotalPrice"].sum()
```

```python
print("Step 8: Monthly Revenue Trend")
print(monthly_sales)
print("\n")
```

```python
df_clean["YearDay"] = df_clean["InvoiceDate"].dt.to_period("D")
day_sales = df_clean.groupby("YearDay")["TotalPrice"].sum()
```

```python
print("Step 8: Daily Revenue Trend")
print(day_sales)
print("\n")
```

# Step 9 : Customer Analysis (RFM style basics)


```python
customer_summary = df_clean.groupby("CustomerID").agg(
    TotalSpent=("TotalPrice", "sum"),
    NumOrders=("InvoiceNo", "nunique"),
    AvgOrdersValue=("TotalPrice", "mean"),
    LastPurchase=("InvoiceDate", "max")
).sort_values("TotalSpent", ascending=False)
```

```python
# show result
print("STEP 9: Top 10 Customers by Total spend")
print(customer_summary.head(5))
print("\n")
```

# Step 10 numpy based statistical analysis on order values

```python
order_values = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().values
```

```python
order_values
```

```python
print("Step 10: Order Value Statistics(Numpy)")
print("Mean:", np.mean(order_values))
print("Median:", np.median(order_values))
print("Standard Deviation:", np.std(order_values))
print("Minimum:", np.min(order_values))
print("Minimum:", np.min(order_values))
print("Maximum:", np.max(order_values))
print("25th percentile:", np.percentile(order_values, 25))
print("75th percentile:", np.percentile(order_values, 75))



```

# Step 11 : Save cleaned data & summary op

```python
df_clean.to_csv("cleaned_Ecommerce_data.csv", index=False)
customer_summary.to_csv("customer_summary.csv")
country_revenue.to_csv("country_revenue.csv")
```

```python
print("Step11: Files Saved -> cleaned_Ecommerce_data.csv, customer_summary.csv, country_revenue.csv")
```

# Step 12 : Data Visualisation (Matplotlb+seaborn)

```python
#Using seaborn to set overall theme
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
plt.figure()
monthly_sales.plot(kind="line", marker="o", color="royalblue")
plt.title("Monthly Revenue Trend", fontsize=14, fontweight="bold")
plt.xlabel("Year-Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=top_products_revenue.values,
    y=top_products_revenue.index,
    palette="viridis"
)

plt.title("Top 10 Products by Revenue", fontsize=14, fontweight="bold")
plt.xlabel("Total Revenue")
plt.ylabel("Product")

plt.tight_layout()
plt.savefig("top_products_revenue.png", dpi=150)
plt.show()
```

```python
#3 top 10 product by quantity sold bar chart 
plt.figure()
sns.barplot(
    x=top_products_quantity.values,
    y=top_products_quantity.index,
    hue=top_products_quantity.index,
    palette="magma"
)
plt.title("Top 10 Products by Quantity Sold", fontsize=14, fontweight="bold")
plt.xlabel("Quantity Sold")
plt.ylabel("Product")

plt.tight_layout()
plt.savefig("top_products_quantity.png", dpi=150)
plt.show()
```

```python
#Revenue by country - top 10 bar chart 
plt.figure()
top_countries = country_revenue.head(10)
sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    hue=top_countries.index,
    palette="crest"
)
plt.title("Top 10 Countries by Revenue", fontsize=14, fontweight="bold")
plt.xlabel("Total Revenue")
plt.ylabel("Country")

plt.tight_layout()
plt.savefig("revenue_by_country.png", dpi=150)
plt.show()
```

```python
#5.Distribution of Order Values (Histogram + KDE)
# ---------------------------------------------------------
plt.figure()
sns.histplot(order_values, bins=50, kde=True, color="teal")
plt.title("Distribution of Order Values", fontsize=14, fontweight="bold")
plt.xlabel("Order Value (Â£)")
plt.ylabel("Frequency")
plt.xlim(0, 1000)  
plt.tight_layout()
plt.savefig("order_value_distribution.png", dpi=150)
plt.show()
```

```python
# 6. Sales by Day of Week (Bar Chart)
#reindex used for handle missing value and outliers
# ---------------------------------------------------------
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday",
             "Friday", "Saturday", "Sunday"]
sales_by_day = (
    df_clean.groupby("DayOfWeek")["TotalPrice"]
    .sum()
    .reindex(day_order)
)
 
plt.figure()
sns.barplot(
    x=sales_by_day.index,
    y=sales_by_day.values,
    hue=sales_by_day.index,
    palette="coolwarm"
)
plt.title("Total Revenue by Day of Week", fontsize=14, fontweight="bold")
plt.xlabel("Day of Week")
plt.ylabel("Total Revenue (Â£)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("sales_by_dayofweek.png", dpi=150)
plt.show()
```

```python
# 7. Sales by Hour of Day (Line Chart)
# ---------------------------------------------------------
sales_by_hour = df_clean.groupby("Hour")["TotalPrice"].sum()
 
plt.figure()
sns.lineplot(x=sales_by_hour.index, y=sales_by_hour.values,
             marker="o", color="darkorange")
plt.title("Total Revenue by Hour of Day", fontsize=14, fontweight="bold")
plt.xlabel("Hour (24-hr format)")
plt.ylabel("Total Revenue (Â£)")
plt.tight_layout()
plt.savefig("sales_by_hour.png", dpi=150)
plt.show()
```

```python
# 8. Top 10 Customers by Total Spend (Bar Chart)
# ---------------------------------------------------------
top_customers = customer_summary.head(10)
 
plt.figure()
sns.barplot(
    x=top_customers["TotalSpent"].values,
    y=top_customers.index.astype(str),
    hue=top_customers.index.astype(str),
    palette="flare"
)
plt.title("Top 10 Customers by Total Spend", fontsize=14, fontweight="bold")
plt.xlabel("Total Spent (Â£)")
plt.ylabel("Customer ID")
plt.tight_layout()
plt.savefig("top_customers.png", dpi=150)
plt.show()
```

```python
#9. Correlation Heatmap (Quantity, UnitPrice, TotalPrice)
# ---------------------------------------------------------
plt.figure()
corr = df_clean[["Quantity", "UnitPrice", "TotalPrice"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.savefig("correlation_heatmap.png", dpi=150)
plt.show()
 
print("STEP 12: All visualizations created and saved as PNG files.")
```
